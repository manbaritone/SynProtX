```python
sample_info = pd.read_csv(f"{RAW_DATA_PATH}/sample_info.csv")
name2depmap = sample_info.iloc[:, 0:2]
dict_annot = dict(zip(name2depmap["cell_line_name"], name2depmap["DepMap_ID"]))
dict_annot
```

```python
df = data_cancer
df["ACH_anno"] = df["cell_line_name"].map(dict_annot)
df = df.loc[pd.isna(df["ACH_anno"])]
df["cell_line_name"].unique()
```

```python
df = data_cancer.copy()

df["ACH_anno"] = df["cell_line_name"].map(dict_annot)
df = df.dropna().reset_index(drop=True)
data_annotated = df.copy()
data_annotated
```

```python
df = pd.read_csv(f"{EXPORT_DATA_PATH}/CCLE_expression_cleaned.csv")
df = df.rename({"Unnamed: 0": "ACH"}, axis=1)
expression_df = df.copy()

expression_col = np.array(expression_df.columns[1:])

expression_df
```

```python
protein_map = pd.read_csv(f"{EXPORT_DATA_PATH}/protein_omics_data_cleaned.csv")
protein_map = protein_map.rename({"Project_Identifier": "cell_line"}, axis=1)
protein_col = np.array(protein_map.columns[1:])
protein_map
```

```python
print("expression_col:", expression_col)
print("protein_col:", protein_col)
```

```python
df = data_annotated.copy()
data_merged = (
    df.merge(expression_df, left_on="ACH_anno", right_on="ACH")
    .merge(protein_map, left_on="cell_line_name", right_on="cell_line")
    .reset_index()
    .rename({"index": "id"}, axis=1)
)
data_merged
```

```python
expression_col = np.append(["id", "cell_line_name"], expression_col)
data_gene = data_merged[expression_col]
data_gene.to_pickle(f"{EXPORT_DATA_PATH}/data_preprocessing_gene.pkl")

data_gene
```

```python
data_gene.nunique()
```

```python
protein_col = np.append(["id", "cell_line_name"], protein_col)
data_protein = data_merged[protein_col]
data_protein.to_pickle(f"{EXPORT_DATA_PATH}/data_preprocessing_protein.pkl")
data_protein
```

```python
drugcomb_col = np.array(data_merged.columns)
drugcomb_col = np.setdiff1d(drugcomb_col, protein_col)
drugcomb_col = np.setdiff1d(drugcomb_col, expression_col)
drugcomb_col = np.setdiff1d(drugcomb_col, ["ACH_anno"])
drugcomb_col = np.append(["id", "cell_line_name"], drugcomb_col)
data_drugcomb = data_merged[drugcomb_col]
data_drugcomb.to_pickle(f"{EXPORT_DATA_PATH}/data_drugcomb.pkl")
data_drugcomb
```
